```python
data1 = []

for page in range(1, 3): #only need 2 pages when selected to view 50000 rows
    url = f"https://www.lpi.usra.edu/meteor/metbull.cfm?sea=*&ants=&nwas=&falls=&valids=&stype=contains&lrec=50000&map=ll&browse=&country=All&srt=name&categ=All&mblist=All&rect=&phot=no&strewn=no&snew=0&pnt=Normal+table&sfor=names&dr=&page={page}"

    r = requests.get(url, headers=headers)

    print(r.text[:1000]) # test print

    soup = BeautifulSoup(r.text, "html.parser")

    # Find the main table
    table = soup.find("table", {"id": "maintable"})

    # Parse table rows (excluding header)
    rows = table.find_all("tr")[1:]  # Skip header

# Extract columns

    for row in rows:
        cols = row.find_all("td")
        if len(cols) == 0:
            continue
        name = re.sub(r'\s+', ' ', cols[0].text.replace("**", "")).strip() # Was having issues with empty spaces and ** denoting and index 
        abbrev = cols[1].text.strip()
        status = cols[2].text.strip()
        fall = cols[3].text.strip()
        year = cols[4].text.strip()
        place = re.sub(r'\s+', ' ', cols[5].text).strip() # Was having issues with empty spaces
        mtype = cols[6].text.strip()
        mass = cols[7].text.strip()
        metbull = cols[8].text.strip()
        latlong = cols[10].text.strip()
   
        data1.append({
            "Name": name,
            "Abbrev": abbrev,
            "Status": status,
            "Fall": fall,
            "Year": year,
            "Place": place,
            "Type": mtype,
            "Mass": mass,
            "MetBull": metbull,
            "LatLong": latlong,
        
        })
```

```python
def clean_year(value):
    if pd.isna(value):
        return np.nan
    
    # Convert to string for processing
    val_str = str(value).strip()
    
    # Handle ranges like "2000 or 2001"
    if 'or' in val_str:
        first_part = val_str.split('or')[0].strip()
        # If first part is a valid year number, return it
        if first_part.isdigit():
            return int(first_part)
        else:
            return np.nan
    
    # If string ends with any non-digit (like letters), mark as NaN
    if not val_str.isdigit():
        return np.nan
    
    # Otherwise, return as integer year
    return int(val_str)

merged_df['Year_clean'] = merged_df['Year'].apply(clean_year)
```

```python
def simplified_type(type_str):
    if pd.isna(type_str):
        return 'Unclassified'
    
    type_str = type_str.strip()

    # Lunar and Martian meteorites
    if 'Lunar' in type_str:
        return 'Lunar'
    if 'Martian' in type_str or any(x in type_str for x in ['Shergottite', 'Nakhlite', 'Chassignite', 'ALH84001']):
        return 'Martian'

    # for other strings
    if any(x in type_str for x in ['Unknown', 'Doubtful', 'Terrestrial', 'Pseudometeorite', 'Discredited', 'Fusion', 'Relict']):
        return 'Unclassified'

    # Iron meteorites - magmatic groups
    iron_magmatic_groups = ['IC', 'IIAB', 'IIC', 'IID', 'IIF', 'IIG', 'IIIAB', 'IIIE', 'IIIF', 'IVA', 'IVB']
    for group in iron_magmatic_groups:
        if type_str.endswith(group):
            return f'{group} iron'

    # Iron meteorites - non-magmatic groups
    if type_str.endswith('IAB'):
        return 'IAB iron'
    if type_str.endswith('IIE'):
        return 'IIE iron'

    # Achondrites
    if any(x in type_str for x in ['Acapulcoite', 'Lodranite', 'Winonaite']):
        if 'Acapulcoite' in type_str:
            return 'ACA achondrite'
        if 'Lodranite' in type_str:
            return 'LOD achondrite'
        if 'Winonaite' in type_str:
            return 'WIN achondrite'
    
    if any(x in type_str for x in ['Howardite', 'Eucrite', 'Diogenite']):
        if 'Howardite' in type_str:
            return 'HOW achondrite'
        if 'Eucrite' in type_str:
            return 'EUC achondrite'
        if 'Diogenite' in type_str:
            return 'DIO achondrite'
    if 'Angrite' in type_str:
        return 'AND achondrite'
    if 'Aubrite' in type_str:
        return 'AUB achondrite'
    if 'Ureilite' in type_str:
        return 'Ureilite achondrite'
    if 'Brachinite' in type_str:
        return 'Brachinite achondrite'
    
    # Stony-iron meteorites
    if 'Pallasite' in type_str:
        return 'PAL'
    if 'Mesosiderite' in type_str:
        return 'MES'  
    
    # Chondrites
    #if 'Chondrite' in type_str or 'chondrite' in type_str or 'chondrites' in type_str:
        # Carbonaceous chondrites and subclasses
    if type_str.startswith('CI'):
        return 'CI chondrite'
    if type_str.startswith(('CM')):
        return 'CM chondrite'
    if type_str.startswith('CO'):
        return 'CO chondrite'
    if type_str.startswith('CV'):
        return 'CV chondrite'
    if type_str.startswith('CK'):
        return 'CK chondrite'
    if type_str.startswith('CR'):
        return 'CR chondrite'
    if type_str.startswith('CH'):
        return 'CH chondrite'
    if type_str.startswith('CB'):
        return 'CB chondrite'

        # Ordinary chondrites
    if type_str.startswith('LL'):
        return 'LL chondrite'
    if type_str.startswith('L'):
        return 'L chondrite'
    if type_str.startswith('H'):
        return 'H chondrite'
        
    # Enstatite chondrites
    if type_str.startswith('EH'):
        return 'EH chondrite'
    if type_str.startswith('EL'):
        return 'EL chondrite'
        
        # Other chondrite groups
    if type_str.startswith('R'):
        return 'R chondrite'
    if type_str.startswith('K'):
        return 'K chondrite'
    if 'Chondrite' in type_str or 'chondrite' in type_str or 'chondrites' in type_str:
        return 'Other chondrite'
    
    # Catch-all for Iron Meteorites if 'Iron' present
    if 'Iron' in type_str:
        return 'Other iron'

    # Unclassified or doubtful cases

    return 'Unclassified'
```

```python
def convert_mass(value):
    if pd.isna(value):
        return np.nan
    value = value.strip().lower()
    if value.endswith("kg"):
        try:
            return float(value.replace("kg", "").strip()) * 1000  # convert kg values to grams
        except:
            return np.nan
    elif value.endswith("g"):
        try:
            return float(value.replace("g", "").strip())  # remove any srings for already in grams
        except:
            return np.nan
     
    elif re.fullmatch(r"\d+(\.\d+)?", value): # if values have no units - assume grams
        try:
            return float(value)
        except:
            return np.nan 
    
    return np.nan

merged_df["Mass_g"] = merged_df["Mass"].apply(convert_mass)
merged_df.drop("Mass", axis=1, inplace=True)
```

```python
# Try to get all pieces in numerical

pieces_map = {
    'one': 1,
    'many': 50,
    'Many': 50,
    'several': 20,
    'Several': 20,
    'hundreds': 200,
    'Hundreds': 200,
    'numerous': 40,
    'frags.': 10,
    'Many fragments': 50,
}

def parse_pieces(val):
    val = str(val).strip().lower()
    
    # Handle fuzzy words
    if val in pieces_map:
        return pieces_map[val]
    
    # Handle different phrases ">100", "over 20", "~30"
    if re.match(r'^>\d+', val):
        return float(re.findall(r'\d+', val)[0]) + 1
    if re.match(r'^~\d+', val):
        return float(re.findall(r'\d+', val)[0])
    if 'over' in val:
        return float(re.findall(r'\d+', val)[0]) + 1
    
    # Extract the first number found
    match = re.search(r'\d+', val)
    if match:
        return float(match.group())
    
    return np.nan  # If nothing found

merged_df["pieces_numeric"] = merged_df["Pieces"].apply(parse_pieces)
merged_df.drop("Pieces", axis=1, inplace=True)
```

```python
def clean_shock(val):
    val = str(val).strip().lower()

    #this was creating new data that I didn't want
    if val in ['0', '7']:
        return np.nan

    # Custom word-based mappings
    shock_map = {
        'low': 'S2',
        'moderate': 'S4',
        'high': 'S5',
        'medium': 'S3',
        'minimal': 'S1'
    }

    for word, mapped in shock_map.items():
        if word in val:
            return mapped

    # Match direct or compound S-levels
    match = re.findall(r'\bS[1-6]\b', val.upper())
    if match:
        return match[0]
    
    return np.nan


merged_df["clean_shock"] = merged_df["Shock"].apply(clean_shock)
merged_df.drop("Shock", axis=1, inplace=True)
```

```python
def clean_weath(val):
    val = str(val).strip().lower()

    # Remove null-like entries (e.g. blank, dash, etc.)
    if val in ['', '-', 'none']:
        return np.nan

    # Word-based mapping
    weath_map = {
        'low': 'W2',
        'very low': 'W1',
        'moderate': 'W3',
        'medium': 'W4',
        'mod': 'W3',
        'minimal': 'W1',
        'minor': 'W1',
        'mild': 'W2',
        'high': 'W5',
        'strong': 'W5',
    }

    for word, mapped in weath_map.items():
        if word in val:
            return mapped

    # Match W1–W6 (including variants like W1-2, W3/4, W4.0, etc.)
    match = re.findall(r'\bW[1-6]\b', val.upper())
    if match:
        return match[0]

    return np.nan


merged_df["clean_weath"] = merged_df["Weath"].apply(clean_weath)
merged_df.drop("Weath", axis=1, inplace=True)
```

```python
# Drop rows with missing or malformed LatLong
latlong_clean = (
    merged_df['LatLong']
    .dropna()
    .str.replace(r'[\(\)]', '', regex=True)
    .str.strip()
)

latlong_clean = latlong_clean[latlong_clean != '']


# Split non-null LatLong values
latlong_split = latlong_clean.str.split(',', expand=True)
latlong_split.columns = ['Latitude', 'Longitude']

# Convert to float
latlong_split['Latitude'] = latlong_split['Latitude'].astype(float)
latlong_split['Longitude'] = latlong_split['Longitude'].astype(float)

# Assign back to the original DataFrame (will align indexes properly)
merged_df['Latitude'] = latlong_split['Latitude']
merged_df['Longitude'] = latlong_split['Longitude']
```
